```python
df = spark.read \
    .format("csv") \
    .option("header", "true") \
    .option("inferSchema", "true") \
    .option("quote", '"') \
    .option("escape", '"') \
    .load("Files/athlete_events.csv")
display(df)
```

```python
df.printSchema()
df.show(5)
```

```python
# Remove surrounding quotes from all column names
df = df.toDF(*[c.strip('"') for c in df.columns])

# Now write safely
df.write.format("delta").mode("overwrite").option("overwriteSchema", "true").saveAsTable("athlete_events")
```
